**Цель исследования:**

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах.

```python
def to_snake_case(name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name) 
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower()
```

Все довольно "ровненько". Можно выделить февраль, как лидера по заключаемым контрактам, но как позже мы увидим - это связано с февралем 2014 года, когда компания и начала свою  активную деятельность.

```python
active_contract = pd.DataFrame()

active_contract['begin_month'] = contract['begin_date'].dt.to_period('M').dt.to_timestamp()
active_contract['end_month'] = contract['end_date'].dt.to_period('M').dt.to_timestamp()

date_range = pd.date_range(start=active_contract['begin_month'].min(), end=active_contract['end_month'].max(), freq='MS')

begin_counts = active_contract.groupby('begin_month').size()
end_counts = active_contract.groupby('end_month').size()

active_contracts = pd.DataFrame(date_range, columns=['month'])
active_contracts.set_index('month', inplace=True)

active_contracts['begin_count'] = begin_counts
active_contracts['end_count'] = end_counts

active_contracts.fillna(0, inplace=True)

active_contracts['active_contracts'] = active_contracts['begin_count'].cumsum() - active_contracts['end_count'].cumsum()

plt.figure(figsize=(14, 10))

sns.lineplot(data=active_contracts, x=active_contracts.index, y='active_contracts')
plt.title('Активные контракты ')
plt.xlabel('Дата')
plt.ylabel('Количество')
plt.xticks(rotation=45)
plt.show()

active_contracts.head(10)
```

- Видно, что активный прирост пользователей начался с 2014-02-01
- в ноябре 2020 тоже заметен большой прирост пользователей - возможно была некая маркетинговая активность
- Good News - Количество активных контрактов стабильно растет - интересно, что с прибылью компании.

**Даты begin_date и end_date**
- Наши данные охватывают период с 2013-10-01 по 2020-01-01

Половина пользователей стримит ТВ и половина пользовотелей стримит фильмы, остается понять - это одни и те же люди или нет. В этом нам поможет корреляционный анализ

```python
# Воспользуемся корреляцией Phik

phik_overview = internet.drop(['customer_id'], axis=1).phik_matrix()
phik_overview.round(2)

plot_correlation_matrix(phik_overview.values, 
                        x_labels=phik_overview.columns, 
                        y_labels=phik_overview.index, 
                        vmin=0, vmax=1, color_map="Blues", 
                        title=r"Корреляция $\phi_K$", 
                        fontsize_factor=1.2, 
                        figsize=(14, 12))
plt.tight_layout()
plt.show()
```

1. Все признаки имеют по две категории

2. Корреляционный анализ с использованием φ<sub>k</sub> показывает, что многие интернет услуги, такие как 
   - стриминг фильмов и ТВ
   - техподдержка и защита устройств

    имеют значительную положительную корреляцию. Это означает, что пользователи, подписанные на одну из этих услуг, часто подписаны и на другие. Эти знания могут быть полезны для принятия бизнес-решений, направленных на перекрестные продажи и маркетинг связанных услуг и, конечно же,  на удержание клиента.

```python

def plot_stacked_bar_seaborn(data, feature, target='target'):
    df = data.copy()
    df['count'] = 1
    df_grouped = df.groupby([feature, target]).count().reset_index()
    df_pivot = df_grouped.pivot(index=feature, columns=target, values='count')
    df_pivot = df_pivot.div(df_pivot.sum(axis=1), axis=0) * 100 
    print(f'Распределение целевого признака по {feature.capitalize()} в %', '\n')
    print(df_pivot.round(1))
    df_pivot.plot(kind='bar', stacked=True, colormap='Set3', figsize=(10, 6))
    plt.title(f'Распределение целевого признака по {feature.capitalize()}')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Percentage')
    plt.xticks(rotation=45)
    plt.legend(title='Target')
    plt.show()

# Построение диаграмм для каждого категориального признака
for feature in categorical_features:
    plot_stacked_bar_seaborn(data, feature)
```

# Подготовка данных для обучения модели

```python
RANDOM_STATE = 60524
TEST_SIZE = 0.25
```

```python
data.shape
```

Разделим данные на тренировочные и тестовые

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop(['target'], axis=1),
    data['target'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE)

X_train.shape, X_test.shape
```

не используем  stratify = data['target']
При стратификации получали незначительное увелечение ROC AUC на кроссвалидации, но на тестовой результат оказывается хуже + увеличивается количество ошибок первого рода

Создадим pipeline предобработки признаков

```python
ohe_columns = [col for col in X_train.columns if X_train[col].dtype == 'object']
ohe_columns
```

```python
num_columns = [col for col in X_train.columns if X_train[col].dtype != 'object']
num_columns
```

```python
ohe_pipe = Pipeline([('ohe', OneHotEncoder(drop='first', handle_unknown='ignore'))])
```

```python
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

Зададим диапазоны гиперпараметров для двух моделей DecisionTreeClassifier() и LGBMClassifier()

```python
param_distributions = [

    # модель DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 70),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # модель LGBMClassifier()
    {
        'models': [LGBMClassifier(random_state=RANDOM_STATE, verbose=0)],
        'models__num_leaves': range(30, 60, 2),  
        'models__max_depth': range(10, 20),
        'models__num_threads':[1],
        'preprocessor__num': [StandardScaler()]
    }    
]
```

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

# Обучение моделей машинного обучения

## DecisionTreeClassifier и LGBMClassifier

Обучим и подберем лучшие параметры для DecisionTreeClassifier() и LGBMClassifier() с помощью RandomizedSearchCV

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_distributions=param_distributions, 
    scoring=('roc_auc'), 
    random_state=RANDOM_STATE,
    cv=5,
    n_iter=10,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)
```

```python
results_df = pd.DataFrame(randomized_search.cv_results_)

best_models = {}

for i, row in results_df.iterrows():
    model_name = row['param_models'].__class__.__name__
    mean_roc_auc = row['mean_test_score']

    if model_name not in best_models:
        best_models[model_name] = {
            'mean_roc_auc': mean_roc_auc,
            'params': row['params']
        }
```

Сохраним лучшую модель 

```python
best_lgbm = randomized_search.best_estimator_
best_lgbm
```
